## Introduction

Canada has become one of the leading Western nations worldwide to acquire migrants in proportion to its overall population. As a result of urbanization and increase in population projections, it has become important to examine the growth of the city quantitatively in order to take advantage of urban opprtunities. As the number of migrants into Western nations continues to expand exponentially, it becomes vital for one to choose a neighbourhood with facilities that tailor to an individual's sense of belonging within a community by making smart and efficient decisions. 

The aim of this project is to analyze the characteristics of an ideal neighborhood of people moving to the multicultural hub of Scarborough, Toronto in comparison to other neighborhoods. These characteristics include average housing price, educational facilities, crime rates, road connectivity, weather conditions, medical care/emergency facilities, fresh water resources, waste management, and recreational facilities.

A quantitative analysis of these features for creating an individual's choice of community may allow further analysis of the urban sector such as:

- Classification of neighborhoods based on socio-economic data 
- To develop and explore urban footprint expansion particularly in underdeveloped neighborhoods
- Identifying urban businnes oppurtunities 
- To understand the intersection of multiculturalism in comparison to racially distinct communities

This project will focus on the location of Scarborough, Toronto as it is a popular destination for new immigrants to reside in Canada. As a result, Scarborough is considered to be a multicultural hub for numerous cultural/ethical practices and landmarks in Metropolitan Toronto. Contrary to recent topics of immigration restrictions, the general trend of immigration into Canada continues to rise despite of this.

## Foursquare API

In order to acquire data on different venues of different neighborhoods, foursquare locational information will be used. Foursquare is a location data provider with information about all manner of venues and events within an area of interest. Such information includes venue names, locations, menus and even photos. As such, the foursquare location platform will be used as the sole data source since all the stated required information can be obtained through the API. We will connect the Foursquare API to the list of neighborhoods in order to gather information about venues inside each and every neighborhood. For each neighborhood, we have chosen the radius to be 100 meter and radius parameter to 500. 

The data retrieved from Foursquare contained information of venues within a specified distance of the longitude and latitude of the postcodes. The information obtained per venue are as follows:

-  Neighborhood
-  Neighborhood Latitude
-  Neighborhood Longitude
-  Venue
-  Name of the venue e.g. the name of a store or restaurant
-  Venue Latitude
-  Venue Longitude
-  Venue Category
